```python
#import matplotlib.lines as mlines
#import matplotlib.transforms as mtransforms

# get plotting data
x_hk = pollution_final["date"].where(pollution_final["city"] == "hong kong")
y_hk = pollution_final["NO_2"].where(pollution_final["city"] == "hong kong")

x_ber = pollution_final["date"].where(pollution_final["city"] == "berlin")
y_ber = pollution_final["NO_2"].where(pollution_final["city"] == "berlin")

x_mad = pollution_final["date"].where(pollution_final["city"] == "madrid")
y_mad = pollution_final["NO_2"].where(pollution_final["city"] == "madrid")

# define axis and parameters
x = (x_hk, x_ber, x_mad)
y = (y_hk, y_ber, y_mad)
colors = ("tab:orange", "blue", "green")
groups = ("HK", "Ber", "Mad")

# create plot
fig = plt.figure(figsize=(10, 6), dpi=80, frameon=True)
ax = fig.add_subplot(1, 1, 1)
ay = fig

# set labels and layout
ax.set_ylabel("g / m3", fontsize=12)
#set_xlabel("Date", fontsize=12)
plt.grid(True)
#plt.tight_layout()  # delete frame
#plt.plot([40], 'k-', lw=2)

for x, y, color, group in zip(x, y, colors, groups):
    ax.scatter(x, y, alpha=0.5, c=color, edgecolors='none', s=20, label=group)
# alpha: set intensity, s: dotsize
ax.axhline(y=40, xmin=0, xmax=1,linewidth=1.5, color="r", ls="--")

#plt.title('Matplot scatter plot')
plt.legend(loc=2)
plt.show()
```
