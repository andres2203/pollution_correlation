# Merging all pollution dataframes

- retrieve stored pickle files for pollution data from all cities

```python
poll_ber = pd.read_pickle('./pickles/berlin_polluion.pkl')
poll_ber
```

```python
poll_hk = pd.read_pickle('././pickles/hk_pollution.pkl')
poll_hk
```

```python
poll_mad = pd.read_pickle('./pickles/madrid_pollution.pkl')
poll_mad
```

```python
all_pollution=pd.concat([poll_ber,poll_hk,poll_mad],ignore_index=True)
all_pollution
```

```python
# Creating new column
all_pollution["date"] = pd.to_datetime(all_pollution["date"], infer_datetime_format=True)
all_pollution.dtypes
```

```python
all_pollution
```

### Creating new column for the year

```python
pollution_final=all_pollution.copy()
```

```python
pollution_final=pollution_final.assign(year=pollution_final['date'].dt.year)
```

```python
pollution_final['city'].value_counts()
```

```python
pollution_final.to_pickle('./pickles/merged_pollution.pkl')
```
